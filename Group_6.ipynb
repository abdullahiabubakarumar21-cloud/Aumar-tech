```python
df = df.dropna()
df.isnull().sum()
```

```python
def convert_sqft_to_num(x):
    # If x is already a number (int or float), return it directly.
    if isinstance(x, (int, float)):
        return float(x)

    # Otherwise, treat it as a string for further processing.
    s = str(x).strip() # Convert to string and remove leading/trailing whitespace

    tokens = s.split('-')
    if len(tokens) == 2:
        try:
            return (float(tokens[0]) + float(tokens[1])) / 2
        except ValueError:
            return None # Handle cases like 'abc-def'
    else:
        try:
            return float(s)
        except ValueError:
            return None # Handle cases like 'invalid'

df['total_sqft'] = df['total_sqft'].apply(convert_sqft_to_num)
df = df.dropna()
```

```python
df['price_per_sqft'] = df['price'] * 100000 / df['total_sqft']
df.head()
```
